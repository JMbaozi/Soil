```python
import arcpy

arcpy.env.workspace = r"E:\GIS\test.gdb"
arcpy.env.overwriteOutput = True

# 输出随机点
output_fc = "soil_random_points"

arcpy.management.CreateRandomPoints(
    out_path=arcpy.env.workspace,
    out_name=output_fc,
    constraining_feature_class=r"E:\GIS\area_boundary.shp",  # 约束范围
    number_of_points_or_field=100,  # 随机点数量
)
print("随机点生成成功：", output_fc)
```
